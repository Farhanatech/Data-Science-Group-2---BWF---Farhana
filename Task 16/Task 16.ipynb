```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]

plt.plot(x,y)
plt.title('Simple Line plot')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.show()
```

```python
y1=[2,3,5,7,11]
y2=[1,4,6,8,10]

plt.plot(x,y1,label='Line 1')
plt.plot(x,y2,label='Line 2')
plt.title('Multiple Lines')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.legend()
plt.show()

```

```python
categories =['A','B','C']
values=[3,7,5]
plt.bar(categories,values,color='green')
plt.title('Vertical Bar Chart')
plt.xlabel('category')
plt.ylabel('value')
plt.show()
```

```python
categories=['A','B','C']
values=[3,7,5]

plt.bar(categories,values,color='grey')
plt.title('Vertical Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
fig,axs=plt.subplots(2,2)
axs[0,0].plot(x,y,color='green')
axs[0,0].set_title('Line plot')

axs[0,1].scatter(x,y)
axs[0,1].set_title('Scatter Plot')

axs[1, 0].bar(categories, values)
axs[1, 0].set_title('Bar Chart')

data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
axs[1, 1].hist(data, bins=4,color='pink')
axs[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5, 6]

plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```
