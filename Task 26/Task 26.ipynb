```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv('dataset_heart.csv')
```

```python
df.head()
```

```python
x=df.drop(columns='heart disease')
y=df['heart disease']
```

```python
y.head()
```

# splitting data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

# logistic regression

```python
model=LogisticRegression(max_iter=1000)
model.fit(x,y)
```

```python
probabilities=model.predict_proba(x)
```

```python
probabilities[:5]
```

```python
predicted_classes = model.predict(x)
print(predicted_classes[:5])
```

# k-Nearest Neighbors (k-NN)

```python
k=5
model=KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
```

```python
print(model.predict(x_test)) 
```

# Support Vector Machines (SVM)

```python
model=SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.predict(x_test[:5])
```

# Decision Trees

```python
tree=DecisionTreeClassifier(criterion='gini',)
```

```python
tree.fit(x_train,y_train)
```

```python
tree.predict(x_test)
```

# Random Forests

```python
rf_random=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
rf_random.fit(x_train,y_train)
```

```python
y_pred=rf_random.predict(x_test)
```

```python
y_pred
```

# Gradient Boosting

```python
boost_model=GradientBoostingClassifier(n_estimators=100,random_state=42)
```

```python
boost_model.fit(x_train,y_train)
```

```python
y_predi=boost_model.predict(x_test)
```

```python
y_predi
```

# Naive Bayes

```python
n_model=GaussianNB()
n_model.fit(x_train,y_train)
```

```python
y_predict=n_model.predict(x_test)
```
