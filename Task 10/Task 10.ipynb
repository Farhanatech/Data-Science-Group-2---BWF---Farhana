```python
array3=np.arange(10)
r1=np.mean(array3)
print("mean",r1)

r2=np.std(array3)
print("Standard Deviation: ",r2)

r3=np.median(array3)
print("Median: ",r3)
```

```python
arr=np.array([[1,2,3],[4,5,7],[8,7,55]])
max_value=np.max(arr)
print("maximum Number: ",max_value)

min_value=np.min(arr)
print("Minimum Number: ",min_value)
```
