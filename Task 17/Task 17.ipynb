```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [25, 16, 9, 4, 1]
categories = ['A','B','C']
values = [5,7,3]
data = [1,2,2,3,3,3,4,4,4,4,5]

fig,axs=plt.subplots(2,2,figsize=(10,10))

axs[0,0].plot(x,y1)
axs[0,0].set_title('Line Plot')

axs[0,1].scatter(x,y2,color='r')
axs[0,1].set_title('scatter plot')

axs[1,0].bar(categories,values,color='g')
axs[1,0].set_title('Bar plot')

axs[1,1].hist(data, bins=5,color='b',alpha=0.7)
axs[1,1].set_title('histogram')

plt.tight_layout()
plt.show()
```

```python
d1 = [1,2,2,3,3,3,4,4,4,4]
d2 = [2,3,3,4,4,4,5,5,5,5]

plt.hist(d1,bins=5,alpha=0.5, label='Data 1', color='blue',density=True)
plt.hist(d2, bins=5 ,alpha=0.5,label='Data 2',color='red',density=True)

plt.title('multiple histograms')
plt.xlabel('value')
plt.ylabel('density')
plt.legend()
plt.show()
```

```python
d1 = [1,2,2,3,3,3,4,4,4,4]
d2 = [2,3,3,4,4,4,5,5,5,5]
d3 = [1,3,3,4,4,4,4,5,5,5]

plt.boxplot([d1,d2,d3],patch_artist=True,boxprops=dict(facecolor='lightblue',color='blue'),medianprops=dict(color='red'))

plt.title('Multiple Box Plots')
plt.xlabel('dataset')
plt.ylabel('value')
plt.xticks([1,2,3], ['Data 1','Data 2','Data 3'])
plt.show()
```

```python
d1 = [1,2,2,3,3,3,4,4,4,4]
d2 = [2,3,3,4,4,4,5,5,5,5]

plt.hist(d1,bins=5,alpha=0.5, label='nation 1', color='grey',density=True)
plt.hist(d2, bins=5 ,alpha=0.5,label='nation 2',color='yellow',density=True)

plt.title('multiple histograms')
plt.xlabel('age')
plt.ylabel('jobs')
plt.legend()
plt.show()
```
