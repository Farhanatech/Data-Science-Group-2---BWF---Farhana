```python
import pandas as pd
import numpy as np
```

```python
#1
df=pd.read_csv('student-scores.csv')
print("Dataframe: ")


print("First 5 rows")
print(df.head())

print("Last 5 rows")
print(df.tail())
```

```python
print(df.columns)
```

```python
#2
df = df.set_index('first_name')
print(df)
```

```python
#3
column1 = df['chemistry_score']
print(column1)
```

```python
column2 = df['physics_score']
print(column2)
```

```python
#4
select_col = df[['gender', 'career_aspiration']]
print("\nSelected Columns:")
print(select_col)
```

```python
#5
subset_rows = df.loc[['Paul', 'Tara', 'Natalie']]
print("\nSubset of Rows:")
print(subset_rows)

```

```python
#6
sub_rows_col = df.iloc[1:4, 2:6]
print("\nSubset of Rows and Columns using .iloc:")
print(sub_rows_col)
```

```python
#7
row1=df[(df['math_score'] > 85) & (df['physics_score']> 90)]
print("First condition")
print(row1.head())

row2= df[(df['gender'] == 'female') & (df['absence_days']< 5)]
print("2nd condition: ")
print(row2.head())

row3=df[(df['gender']== 'male') & ((df['part_time_job']== True) | (df['absence_days']> 5))]
print("3rd condition: ")
print(row3.head())
```

```python
#8
grouped=df.groupby('gender').mean(numeric_only=True)
print(grouped)
```

```python
#9
groupedvar=df.groupby(['gender', 'part_time_job']).sum()
print(groupedvar)
```

```python
#10
group_df = df.groupby(['gender','part_time_job']).agg({'absence_days': ['mean', 'sum'],
    'math_score': ['mean','max'],
    'physics_score': ['mean','min','max']
})
print(group_df)
```

```python
#11
groupedvar=df.groupby(['gender', 'part_time_job']).size()
print(groupedvar)
```

```python
#12
row = df[(df['gender'] == 'female') & (df['math_score'] > 90)]
print(row.head())
```

```python
#13
row = df.query("gender == 'female' & math_score > 90")
print(row.head())
```

```python
#14
fil=[' Casey','Sandoval']
fil_df=df[df['last_name'].isin(fil)]
print(fil_df)
```

```python
#15
selec_col = ['last_name','math_score', 'physics_score']
renamed_df = df[selec_col].rename(columns={
   'last_name': 'Last Name',
    'math_score': 'Math Marks',
    'physics_score': 'Physics Marks'
})
print(renamed_df)
```
