```python
#!/usr/bin/env python
# coding: utf-8

# In[53]:


product_details = {} 
product_order = []
product_info = ()
categories = set()

def add_product(product_id, name, category, stock):
    product_info = (name, category)
    product_details[product_id] = {'info': product_info, 'stock': stock}
    product_order.append(product_id)
    categories.add(category)

# Function to update stock
def update_stock(product_id, quantity):
    if product_id in product_details:
        product = product_details[product_id]
        product['stock'] += quantity
        name, category = product['info']
        print(f"Updated stock for {product_id}-{name} of {category} category by {quantity}. New stock: {product_details[product_id]['stock']}")
    else:
        print(f"Product ID {product_id} not found.")

# Function to display all products
def display_products():
    for product_id in product_order:
        product = product_details[product_id]
        name, category = product['info']
        stock = product['stock']
        print(f"Product ID: {product_id}, Name: {name}, Category: {category}, Stock: {stock}")

# Function to display products by category
def display_by_category(category):
    print(f"\nProducts in category '{category}':")
    for product_id in product_order:
        product = product_details[product_id]
        name, product_category = product['info']
        stock = product['stock']
        if product_category == category:
            print(f"Product ID: {product_id}, Name: {name}, Stock: {stock}")




# In[54]:


add_product("ID001", "Milk", "Dairy", 80)
add_product("ID002", "Yogurt", "Dairy", 15)
add_product("ID003", "Eggs", "Poultry", 144)
add_product("ID004", "Bun", "Bakery", 25)
add_product("ID005", "Bread", "Bakery", 30)
add_product("ID006", "Cookies", "Bakery", 50)
add_product("ID007", "Cheese", "Dairy", 24)
add_product("ID008", "Cream", "Dairy", 24)
add_product("ID009", "Chicken", "Meat", 70)
add_product("ID010", "Mutton", "Meat", 40)
add_product("ID011", "Beef", "Meat", 40)
add_product("ID012", "Fish", "Meat", 30)


print("\n\nUpdating stock:\n**************")

update_stock("ID000", +10)
update_stock("ID002", -5)
update_stock("ID010", +20)
update_stock("ID006", -15)

print("\n\nAll Products:\n*************")
display_products()

print("\n\nCategory-wise filtered products\n*******************************")
display_by_category("Bakery")
display_by_category("Dairy")
display_by_category("Poultry")
display_by_category("Meat")

```
