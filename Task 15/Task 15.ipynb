```python
df3=pd.read_csv("basic_data_new.csv")
```

```python
#8
pivot_table = pd.pivot_table(df3, values='Sales', index='Category', columns='Sub-category', aggfunc='sum')
print("Pivot Table:")
print(pivot_table)
```

```python
#9
sum_df = df3.groupby(['Category','Sub-category']).sum()
print("Grouped by Sum")
print(sum_df)

mean_df = df3.groupby('Category')['Sales'].mean()
print("By mean:")
print(mean_df)

count_df = df3.groupby('Category').size()
print("by Category:")
print(count_df)
```

```python
#10
multiple_agr = df3.groupby(['Category', 'Sub-category']).agg({'Sales': ['sum', 'mean', 'count'],'Profit': ['sum', 'mean']
})
print("Multiple Aggregation Functions:")
print(multiple_agr)
```

```python
#11
def cal_range(x):
    return x.max() - x.min()
custom = df3.groupby('Category')['Sales'].apply(cal_range)
print("Custom Function")
print(custom)
```
