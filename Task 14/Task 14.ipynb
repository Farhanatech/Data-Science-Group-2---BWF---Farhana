```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('student-scores.csv')
print("Dataframe: ")
df
```

```python
#1
print(df.isnull())
print(df.isnull().sum())
```

```python
#2
droped_rows=df.dropna()
print(droped_rows)
```

```python
#3
droped_col=df.dropna(axis=1)
print(droped_col)
```

```python
#4
filled_val=df.fillna(0)
print(filled_val)
```

```python
#5
ffill=df.fillna(method='ffill')
print(ffill)
bfill=df.fillna(method='bfill')
print(bfill)
```

```python
#6
df_inter=df.interpolate()
print(df_inter)
```

```python
#7
df['math_score']=df['math_score'].astype(float)
print(df.dtypes)
```

```python
#8
df['math_score']=df['math_score'].apply(lambda x: x*1.1)
print(df['math_score'])
```

```python
#9
df['math_score']=(df['math_score']-df['math_score'].min())/ (df['math_score'].max()-df['math_score'].min())
print(df['math_score'])
```

```python
#10
df['math_score']=(df['math_score']-df['math_score'].mean())/df['math_score'].std()
print("standard normailization")
print(df['math_score'])
```

```python
#11
dupli=df.duplicated()
print("Duplicates row: ")
print(df[dupli])
```

```python
#12
drop_dupli=df.drop_duplicates()
print("dropping duplicated rows")
print(drop_dupli)
```

```python
#13
specific_dupli=df.drop_duplicates(subset=['first_name', 'last_name'])
print("dropping duplicate rows based on first_name and last_name:")
print(specific_dupli)
```

```python
#14
df['first_name']=df['first_name'].str.lower()
print("lower case dataframe: ")
print(df['first_name'])
```

```python
#15
df['first_name'] = df['first_name'].str.strip()
print("removing leading and trailing spaces :")
print(df['first_name'])
```

```python
#16
df['first_name']=df['first_name'].str.replace('pual','john')
print("replaced: ")
print(df['first_name'])
```

```python
#17
df['initial']=df['first_name'].str[0]
print("Extract a Substring from Each Value in a Column")
print(df['initial'])
```

```python
#18
birth_dates = pd.Series(['2001-01-01', '2002-02-02', '2003-03-03', '2004-04-04', '2005-05-05'] * (len(df) // 5) + ['2006-06-06'] * (len(df) % 5))
df['birth_date']=birth_dates
df['birth_date']=pd.to_datetime(df['birth_date'])
print("converting birth_date to datetime:")
print(df[['birth_date']].head())
```

```python
#19
df['year']=df['birth_date'].dt.year
df['month']=df['birth_date'].dt.month
df['day']=df['birth_date'].dt.day
print("Extracted year,month,day")
print(df[['year','month','day']])
```

```python
#20
start_date = '2000-01-01'
end_date = '2010-12-31'
fil_df = df[(df['birth_date'] >= start_date)&(df['birth_date'] <= end_date)]
print("filtering rows based on date range:")
print(fil_df)
```

```python
#21
one_hot_encode=pd.get_dummies(df,columns=['gender'])
print("one_hot encoding")
print(one_hot_encode)
```

```python
#22
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['gender']=le.fit_transform(df['gender'])
print("after label encoding:")
print(df['gender'])
```

```python
#23
category ={
    'Doctor' :'Healthcare',
    'Engineer':'Engineering',
    'Teacher':'Education',
    'Lawyer':'Legal'
}
df['career_group']=df['career_aspiration'].map(category)
print(df)
```

```python
#24
df2=pd.read_csv('student-scores.csv')
merg=pd.merge(df,df2,on='id')
print(merg.head())
```

```python
#25
concat_df_ver = pd.concat([df, df2], axis=0)
print("Concatenated DataFrame Vertical:")
print(concat_df_ver)
```

```python
#26
concat_df_hor = pd.concat([df, df2], axis=1)
print("Concatenated DataFrame Horizontal:")
print(concat_df_hor)
```

```python
#27
df['total_score']=df['math_score']+df['physics_score']
print("new column: ")
print(df[['math_score','physics_score','total_score']])
```

```python
#28
df['score_bin']=pd.cut(df['total_score'],bins=5)
print("discretizing total score into bins: ")
print(df['score_bin'])
```

```python
#29
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
feature=poly.fit_transform(df[['math_score','physics_score']])
poly_df=pd.DataFrame(feature,columns=poly.get_feature_names_out(['math_score','physics_score']))
print("polynomial features: ")
print(poly_df)
```
