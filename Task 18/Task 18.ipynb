```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
```

# Line plot

```python
data=sns.load_dataset("iris")
sns.lineplot(x="sepal_length",y="sepal_width",data=data)

plt.title('Iris plant')

sns.set_style("ticks") 
plt.show()
```

# scatter plot with customization

```python
 sns.set(style='whitegrid')
fmri=sns.load_dataset("fmri")
sns.scatterplot(x="timepoint",
                y="signal",
                hue="region",
                style="event",
                data=fmri)
```

```python
 sns.set(style='whitegrid')
fmri=sns.load_dataset("fmri")
sns.scatterplot(x="timepoint",
                y="signal",
                hue="region",
                style="event",
                palette="pink",
                data=fmri)
```

# Line plot with customization

```python
data=sns.load_dataset("tips")
sns.lineplot(x='size',y='total_bill',data=data)

plt.title("Line plot of size vs total bill")
plt.xlabel('size')
plt.ylabel('total bill')
plt.show()
```

```python
data=sns.load_dataset("tips")
sns.lineplot(x='size',y='total_bill',hue='time',data=data)
sns.set(style="darkgrid")
plt.title("Line plot of size vs total bill")
plt.xlabel('size')
plt.ylabel('total bill')
plt.show()
```

```python
data.info()
```

# Bar plot with customization

```python
sns.barplot(x='day',y='total_bill',data=data)
data.sort_values('total_bill')
sns.set_palette("Greens")
sns.set_style("ticks")
plt.title('Bar Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

# Count Plot with Customization 

```python
sns.countplot(x='day',data=data)
plt.title('Count Plot of Days')
sns.set_palette("Greens")
sns.set_style("ticks")
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

# Box Plot with Customization

```python
sns.boxplot(x='day', y='total_bill',data=data)
sns.set_palette("pastel")
sns.set_style("whitegrid")
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```
