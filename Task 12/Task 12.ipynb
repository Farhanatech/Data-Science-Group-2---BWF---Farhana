```python
# 3
ser1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
ser2 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])

print("series 1")
print(ser1)
print("series 2")
print(ser2)

add=ser1 + ser2
print("adition")
print(add)

sub=ser1 - ser2
print("Subtraction")
print(sub)

multi=ser1 * ser2
print("Multiplication")
print(multi)

divide=ser1 / ser2
print("Division")
print(divide)
```
