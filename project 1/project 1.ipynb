```python
def plot_enrollment_trends_year(df):
    enrollment_by_year=df.groupby('year')['enrollment_count'].sum()
    plt.figure(figsize=(10,6))
    plt.plot(enrollment_by_year.index,enrollment_by_year.values, marker='o')
    plt.xlabel('Year')
    plt.ylabel('total enrollment count')
    plt.title('Enrollment trends by year')
    plt.grid(True)
    plt.show()


def caculate_averg_enrollment(df):
    average_enrollment_online=df[df['course_type']=='online']['enrollment_count'].mean()
    average_enrollment_classroom=df[df['course_type']=='classroom']['enrollment_count'].mean()
    return  average_enrollment_online,average_enrollment_classroom

def higher_enrollment(df):
   enrollment_online=df[df['course_type']=='online']['enrollment_count']
   enrollment_classroom=df[df['course_type']=='classroom']['enrollment_count']

   if enrollment_online.mean()>enrollment_classroom.mean():
      return 'online courses have higher enrollment'
   else:
      return 'classroom courses have higher enrollment'
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df[df['course_type'] =='online']['enrollment_count'], bins=20,alpha=0.5, label='Online')
plt.hist(df[df['course_type']=='classroom']['enrollment_count'], bins=20, alpha=0.5,label='Classroom')
plt.xlabel('Enrollment Count')
plt.ylabel('Frequency')
plt.title('Enrollment Distribution by Course Type')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['course_type_encode']=le.fit_transform(df['course_type'])

correl=df['course_type_encode'].corr(df['enrollment_count'])
print("corelation betweeen course type and enrollment_count:")
print(correl)
```

```python
if __name__ == "__main__":
 plot_enrollment_trends_year(df)
 avg_enrollment_online, avg_enrollment_classroom = caculate_averg_enrollment(df)
 print(f"Average Enrollment in Online Courses: {avg_enrollment_online:.2f}")
 print(f"Average Enrollment in Classroom Courses: {avg_enrollment_classroom:.2f}")
    
result =higher_enrollment(df)  
print(result)
```
