```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset("tips")
```

```python
data.head(5)
```

```python
sns.regplot(x='total_bill',y='tip',data=data)
plt.title('Regression plot of tips and total bills')
plt.xlabel('total_bills')
plt.ylabel('tips')
plt.show()
```

```python
g=sns.lmplot(x='total_bill',y='tip',data=data,col='time', hue='day')
g.fig.suptitle('Multiple linear regression plot of tips and total bills',y=1.02)
g.set_axis_labels('Total Bills','Tips')
plt.show()
```

```python
print(data['time'].value_counts())
```

```python
sns.residplot(x='total_bill',y='tip',data=data)
plt.title('Residual plot of tips and total bills')
plt.xlabel('total_bills')
plt.ylabel('residuals')
plt.show()
```

```python
sns.histplot(data['total_bill'], bins=20, kde=False)
plt.title('Histogram of Total Bill Amounts')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], color='skyblue')

plt.title('KDE Plot of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data)
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=data, palette='muted')
plt.title('Violin Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
sns.pairplot(data, hue='day', palette='coolwarm')
plt.suptitle('Pair Plot of Tips Dataset', y=1.02)
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix,annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='scatter', color='purple')
plt.suptitle('Joint Plot of Total Bill and Tip', y=1.02)
plt.show()
```

```python
g = sns.FacetGrid(data, col="time", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip", alpha=.7)
g.add_legend()
plt.suptitle('Facet Grid of Total Bill and Tip by Time and Smoker', y=1.05)
plt.show()
```

```python
sns.distplot(data['total_bill'], bins=30, kde=True, color='skyblue')
plt.title('Distribution Plot of Total Bill Amounts')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
```
